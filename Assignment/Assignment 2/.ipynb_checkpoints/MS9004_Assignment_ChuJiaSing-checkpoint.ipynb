```python
# Define colors for pie charts
colors = ['lightskyblue', 'lightcoral']

# Create the figure and subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 6))

# Create pie charts with titles and counts
for i, col in enumerate(['Genotype', 'Treatment', 'Behavior']):
    counts = df[col].value_counts()
    labels = [f"{l} ({counts[l]})" for l in counts.index]  # Add count in parentheses
    axes[i // 2, i % 2].pie(counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
    axes[i // 2, i % 2].set_title(col, fontsize=12, loc='left')

# Adjust layout and spacing
plt.tight_layout()

# Display the plot
plt.show()
```

### Fit Model

```python
dftrain, dftest = train_test_split(df, test_size=0.25, random_state=80385)
```

```python
mlr1 = ols('Y ~ X1 + X2 + X3 + X4 + X5 + X6 + X7 + X8 + X9 + X10 + Genotype + Treatment + Behavior', dftrain).fit()
print(mlr1.summary())
print("\nMSE =", mlr1.mse_resid)
```

```python
np.round(mlr1.params, decimals=3).tolist()
```

# Evaluate Model

```python
res = mlr1.resid
```

```python
sns.displot(res, kde=True)
```

```python
sm.qqplot(res, line='s')
```

```python
sm.stats.jarque_bera(res)
```

```python
sm.stats.omni_normtest(res)
```

```python
sm.stats.het_breuschpagan(res, mlr1.model.exog)
```

```python
sns.residplot(x=mlr1.fittedvalues, y=res)
```

```python
sm.stats.durbin_watson(res)
```

```python
mlr1.condition_number
```

```python
for i in range(14):
    predictor = mlr1.model.exog_names[i]
    Vif = vif(mlr1.model.exog,i)
    print(predictor, np.round(Vif, decimals=2))
```
